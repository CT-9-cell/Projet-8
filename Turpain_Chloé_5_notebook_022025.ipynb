```python
df_region_country["Country"] = df_population8["Country"].replace("North Macedonia", "Republic of North Macedonia")
```

```python
# Faire un left join entre Région et Population
merged_df = df_region_country.merge(df_population8, on="Country", how="left")

# Vérifier les premières lignes du résultat
print(merged_df.head())
```

```python
# Identifier les lignes non jointes (celles avec des valeurs NaN dans les colonnes de region_df)
non_joined_rows = merged_df[merged_df.isnull().any(axis=1)]
```

```python
# Afficher les lignes non jointes
print(non_joined_rows)
```

```python
merged_df[merged_df.duplicated(subset=["Country", "Year", "Granularity"], keep=False)]
```

```python
# Vérifier les doublons dans les colonnes pivotées
duplicates = merged_df[merged_df.duplicated(subset=["Country", "Year", "Granularity"], keep=False)]
```

```python
merged_df[(merged_df['Country']=='Afghanistan')&(merged_df['Granularity']=='Total')&(merged_df['Year']==2000)]
```

```python
# Focus sur un pays spécifique (exemple : "Afghanistan")
duplicates[duplicates['Country']=='Afghanistan']
```

```python
# Supprimer les doublons tout en gardant la première occurrence
merged_df = merged_df.drop_duplicates(subset=["Country", "Year", "Granularity"])
```

```python
# Pivot pour transformer la granularité en colonnes
pivot_df = merged_df.pivot(index=["Country", "Year"], columns="Granularity", values="Population")
```

```python
# Réinitialiser l'index pour rendre les colonnes plus claires
pivot_df.reset_index(inplace=True)
```

```python
# Afficher le DataFrame pivoté
pivot_df
```

```python
# Effectuer le left join
merged_df_2 = df_drinking_water.merge(merged_df, on=["Country", "Year", "Granularity"], how="left")

# Vérifier les premières lignes du DataFrame joint
merged_df_2
```

```python
# Identifier les lignes non jointes (celles avec des valeurs NaN dans les colonnes de region_df)
non_joined_rows_2 = merged_df_2[merged_df_2.isnull().any(axis=1)]
```

```python
# Afficher les lignes non jointes
print(non_joined_rows_2)
```

```python
# Nombre total de lignes non jointes
print(f"Nombre de lignes non jointes : {len(non_joined_rows_2)}")
```

```python
# Vérifier les doublons dans les colonnes pivotées
duplicates_2 = merged_df_2[merged_df_2.duplicated(subset=["Country", "Year", "Granularity"], keep=False)]
```

```python
# Afficher les lignes dupliquées
print(duplicates_2)
```

```python
# Identifier les pays présents dans le fichier principal mais pas dans le fichier joint
missing_countries = set(merged_df_2["Country"]) - set(df_drinking_water["Country"])
print("Pays non trouvés dans le fichier Drinking Water :", missing_countries)
```

```python
# Identifier les années manquantes
missing_years = set(merged_df_2["Year"]) - set(df_drinking_water["Year"])
print("Années non trouvées :", missing_years)
```

```python
# Identifier les granularités manquantes
missing_granularities = set(merged_df_2["Granularity"]) - set(df_drinking_water["Granularity"])
print("Granularités non trouvées :", missing_granularities)
```
